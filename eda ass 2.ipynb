# Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

```python
flight_prices = df["Price"]

plt.figure(figsize=(10,6))
plt.hist(flight_prices,edgecolor="black",bins=10)
plt.xlabel("flight prices")
plt.ylabel("frequency")
plt.title("distribution of flight prices")
```

# Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
min_price = flight_prices.min()
max_price = flight_prices.max()
print("max price",max_price)
print("min price",min_price)
```

```python
prices_range = max_price-min_price
print("price range",prices_range)
```

Missing values can have a significant impact on your analysis, and their impact depends on the extent and nature of the missing data:

Reduced Sample Size: Missing data reduces the effective sample size, which can lead to a loss of statistical power and potentially affect the reliability of your analysis.

Biased Results: If missing data is not missing at random (MAR) and is related to the outcome or variables of interest, it can introduce bias into your analysis.

Incomplete Insights: Missing values may limit your ability to draw meaningful conclusions and insights from the dataset.

Data Imputation: To address missing data, you may need to perform data imputation techniques such as mean imputation, median imputation, or more advanced methods like regression imputation. The choice of imputation method can impact your analysis results.
